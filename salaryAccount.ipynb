```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Provide the correct path to the CSV file
data = pd.read_csv(r"E:\Python\Data\Data set\Salary_Data.csv")

data.head()
```

```python
x = data["YearsExperience"].values.reshape(-1, 1)
x

y = data["Salary"].values
y
```

```python
# Split Data

xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size=0.7, random_state=0)
```

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
predict_y = model.predict(xtest)
print(predict_y)
```

```python
# Check Accuracy

from sklearn.metrics import r2_score
r2_score(ytest, predict_y)
```

```python
plt.scatter(x, y, color="red")
plt.plot(x, model.predict(x), color="blue")
plt.title('Experience vs Salary (Linear Regression)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
